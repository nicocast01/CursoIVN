### Identificando a una distribucion normal

```python
#Creamos un dataframe de referencia con una distribución normal
np.random.seed(29) # para reproducibilidad

n = 200
x = np.random.randn(n)

data = pd.DataFrame([x]).T
data.columns = ['x']
data.head()
```

```python
sns.histplot(data['x'], color="red", label="100% Equities", kde=True, stat="density", linewidth=0)
```

```python
# Histograma para la variable RM de boston house price dataset
# RM es el número medio de habitaciones por vivienda

sns.histplot(boston['RM'],kde=True, stat="density", linewidth=0)
```

```python
# Hiatograma de la variable LSTAT
# % estado más bajo de la población

sns.histplot(boston['LSTAT'],kde=True, stat="density", linewidth=0, bins=30)
```


Q-Q plots.  

```python
# Grafiquemos la gráfica Q-Q plot para los datos simulados.
# Los puntos deberían ajustarse en la linea de 45 grados

stats.probplot(data['x'], dist="norm", plot=plt)
plt.show()
```

```python
# El mismo procedimiento para RM
stats.probplot(boston['RM'], dist="norm", plot=plt)
plt.show()
```

```python
# para la variable CRIM
stats.probplot(boston['CRIM'], dist="norm", plot=plt)
plt.show()
```

### Visualizando datos atípicos

```python
# Creamos un dataframe con las variables independientes. Usaremos solo 3 variables para esta demo

boston_subset = boston[['RM', 'LSTAT', 'CRIM']] 
boston_subset.head()
```

```python
# boxplot
plt.figure(figsize=(3,6))
sns.boxplot(y=boston_subset['RM'])
plt.title('Boxplot')
```

```python
# boxplot
plt.figure(figsize=(3,6))
sns.boxplot(y=boston_subset['LSTAT'])
plt.title('Boxplot')
```

```python
#scatterplot
sns.set_theme()
#cols = ['SalePrice', 'OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'FullBath', 'YearBuilt']
sns.pairplot(boston_subset)
plt.show()
```

```python
#correlation matrix
corrmat = boston.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True);
```
